```python
# 1. Area of Circle
def area_circle(radius):
    return 3.14159 * radius * radius

print(area_circle(5))
```

```python
# 2. Largest of Three Numbers
def largest(a, b, c):
    return max(a, b, c)

print(largest(10, 25, 15))
```

```python
# 3. String Length Without len()
def string_length(s):
    count = 0
    for char in s:
        count += 1
    return count

print(string_length("Hello"))
```

```python
# 4. Check Prime Number
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(17))
```

```python
# 5. Recursive Fibonacci
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(7))
```

```python
# 6. Recursive String Reverse
def reverse_string(s):
    if len(s) == 0:
        return s
    return reverse_string(s[1:]) + s[0]

print(reverse_string("Hello"))
```

```python
# 7. Sum from 1 to n
n = 10
total = 0

for i in range(1, n + 1):
    total += i

print(total)
```

```python
# 8. Count Even Numbers in a List
numbers = [1, 2, 3, 4, 5, 6, 8]
count = 0

for num in numbers:
    if num % 2 == 0:
        count += 1

print(count)
```

```python
# 9. Print Prime Numbers Between 1 and 100
for num in range(2, 101):
    prime = True

    for i in range(2, num):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num)
```
